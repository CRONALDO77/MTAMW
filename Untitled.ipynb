```python
import matplotlib.pyplot as plt
import numpy as np
"""
datatype = 'mosi_'
labels = ['# 1', '# 2', '# 3']
text = [0.4855,0.5456,0.5622]
visual = [0.2956,0.2359,0.2290]
acoustic = [0.2189,0.2185,0.2089]
"""
"""
datatype = 'speechbrain_'
labels = ['# 1', '# 2', '# 3']
text = [0.4909 ,0.5040,0.5140]
visual = [0.2759,0.2656,0.2555]
acoustic = [0.2332,0.2304,0.2305]
"""

datatype = 'ibm_'
labels = ['# 1', '# 2', '# 3']
text = [0.4870,0.4925,0.4996]
visual = [0.2740,0.2698,0.2631]
acoustic = [0.2390,0.2377,0.2373]



x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(5, 3.5))
rects1 = ax.bar(x - width, text, width, label='Text',color='#B7DEE8')
rects2 = ax.bar(x , visual, width, label='Visual',color='#FEE599')
rects3 = ax.bar(x + width ,acoustic, width, label='Acoustic',color='#D7E3BF')
ax.set_ylim(0.18,0.53)
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Relative Importance')
ax.set_title('MOSI-IBM')
ax.set_xticks(x, labels)
ax.legend(fontsize='x-small',frameon=0,loc='upper right')

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)

fig.tight_layout()
plt.savefig(datatype+'all.pdf')
plt.show()
```
